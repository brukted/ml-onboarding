# Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("imdb_top_1000.csv")
summary = [[df.columns.size, df['IMDB_Rating'].mean(), df['Genre'].nunique()]]

result = pd.DataFrame(data = summary, columns = ["Columns", "Average Rating", "Unique Genres"])
print(result)
result.to_csv("result.csv")
```
